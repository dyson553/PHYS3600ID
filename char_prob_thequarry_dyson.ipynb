```python
import json
```

```python
# Load json pair list (900)
with open('pair_list.json') as f:
  pair_list = json.load(f)
```

```python
# Load json sub-string list (30)
with open('sub_list.json') as f:
  sub_list = json.load(f)
```

### Draft For Loop ###

```python
# Nested for loop that only stores the matches from the sub-strings.

result = []    

for string in pair_list:
    matches = []
    for substring in sub_list:
        if substring in string:
            matches.append(substring)
    if matches:
        result.append(matches)

print(result)
```

### Final Result/Function ###

```python
# Dictionary comprehension where the pair_list is the Key and the matches from the sub-strings are the value.

def char_function(pair_list):
  result_list = {pair_string: "/".join([sub_string 
                      for sub_string in sub_list \
                      if sub_string in pair_string]) \
                      for pair_string in pair_list}

# Display result_list when function is executed
  print(json.dumps(result_list, indent=0))
  
# Save the displayed results_list as a .json file
  with open('matches_data.json', 'w') as fp:
    json.dump(result_list, fp, indent=1)
```
